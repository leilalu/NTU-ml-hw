# 问题描述
## 本次作业采用从中央气象局网站下载的真实观测数据，利用linear regression预测某时刻PM2.5的数值。

## 采用Gradient Descent、Stochastic Gradient Descent、Adagrad等优化算法，并比较它们的性能。


```python
# import
import numpy as np
import csv, os
from numpy.linalg import inv  # numpy.linalg模块包含线性代数的函数。使用这个模块，可以计算逆矩阵、求特征值、解线性方程组以及求解行列式等。在本次作业中主要用于计算参数w的close-form时求解矩阵的逆
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# gradient descent
def GD(X, Y, W, eta, iteration, lambdaL2):
    """
    使用gradient descent时，learning rate要调的很小，不然容易爆炸
    """
    list_cost = [] # 记录下每次迭代的cost function值，便于可视化
    for itera in range(iteration):
        hypo = X.dot(W) # 注意，预测值hypo也是一个向量
        loss = hypo - Y
        cost = np.sum(loss ** 2) / X.shape[0] # shape[0]可以获取矩阵X第0维的长度，即样本数
        list_cost.append(cost)
        
        # 更新参数值
        gradient = X.T.dot(loss) / X.shape[0]  + lambdaL2 * W
        W -= eta * gradient
        # 每1000次迭代，输出一下当前的cost function值，确保没有陷入local minima等
        if itera % 1000 == 0:
            print("iteration : {}, cost : {}".format(itera, cost))
    return W, list_cost
    
```

```python
# stochastic gradient descent
def SGD(X, Y, W, eta, iteration, lambdaL2):
    list_cost = []
    for itera in range(iteration):
        hypo = X.dot(W)
        loss = hypo - Y
        cost = np.sum(loss ** 2) / X.shape[0]
        list_cost.append(cost)
        
        # 更新参数
        
        gradient = X.T.dot(loss) / X.shape[0]  +  lambdaL2 * W
        W -= eta * gradient
        if itera % 1000 == 0 :
            print("iteration:{}, cost:{}".format(itera, cost))
    return W, list_cost
        
```

```python
# Adagrad
def Adagrad(X, Y, W, eta, iteration, lambdaL2):
    list_cost = []
    gradient_sum = np.zeros(X.shape[1])
    for itera in range(iteration):
        hypo = np.dot(X, W)
        loss = hypo - Y
        cost = np.sum(loss ** 2) / X.shape[0]
        list_cost.append(cost)
        
        # 更新参数
        gradient = X.T.dot(loss) / X.shape[0]  + lambdaL2 * W
        gradient_sum += gradient ** 2
        W -= eta * gradient / np.sqrt(gradient_sum)
        
        if itera % 1000 == 0 :
            print("iteration:{}, cost:{}".format(itera, cost))
    return W, list_cost
        
```

### 数据预处理

```python
data = []
for i in range(18):
    data.append([])

# 将资料读取到data中
textTrain = open(os.path.join(os.path.dirname('__file__'), "data_01/train.csv"), "r", encoding="big5")  # os.path.dirname(path)的作用是去掉文件名，返回目录。而os.path.dirname(__file__)可以用来获取当前文件所在路径，从而获取相对路径,当报错【file未定义时】，在file两边加上单引号即可
rowTrain = csv.reader(textTrain)
n_row = 0 # 设为0是因为csv中第一行是表头，不需要读取
for row in rowTrain:
    if n_row != 0: 
        for i in range(3, 27):
            if row[i] != "NR":
                data[(n_row - 1) % 18].append(float(row[i]))
            else:
                data[(n_row - 1) % 18].append(float(0)) # 缺失值补0
    n_row += 1
textTrain.close()        
```

```python
trainX = []
trainY = []
for m in range(12):  # 12 个月
    for i in range(471): # 每个月有20天，每天24个小时，10个小时一笔资料，共有417笔
        trainX.append([])
        trainY.append(data[9][480*m + i + 9]) # data[9]是PM2.5
        # 18种污染物
        for p in range(18):
            # 收集9小时的资料
            for t in range(9):
                trainX[471 * m + i].append(data[p][480*m + i + t])
    
```

```python
testData = []
textTest = open(os.path.join(os.path.dirname('__file__'), "data_01/test.csv"), "r", encoding="big5")
rowTest = csv.reader(textTest)
n_row = 0
for row in rowTest:
    if n_row %18 == 0:
        testData.append([])
        for i in range (2, 11):
            testData[n_row // 18].append(float(row[i]))
    else:
        for i in range(2,11):
            if row[i] == "NR":
                testData[n_row // 18].append(float(0))
            else:
                testData[n_row // 18].append(float(row[i]))
    n_row += 1
textTest.close()

testX = np.array(testData)
trainX = np.array(trainX)
trainY = np.array(trainY)
```

```python
# 增加bias项
trainX = np.concatenate((np.ones((trainX.shape[0], 1)), trainX), axis=1) #(5652, 163)
```

```python
# gradient descent
lr = 1e-6
W = np.zeros(trainX.shape[1]) #(163, )
W_gd, list_cost_gd = GD(X=trainX, Y=trainY, W=W, eta=lr, iteration=20000, lambdaL2=0)

W = np.zeros(trainX.shape[1]) #(163, )
W_gd_1, list_cost_gd_1 = GD(X=trainX, Y=trainY, W=W, eta=lr, iteration=20000, lambdaL2=100)

# Adagrad
lr = 5
W = np.zeros(trainX.shape[1]) #(163, )
W_ada, list_cost_ada = Adagrad(X=trainX, Y=trainY, W=W, eta=lr, iteration=20000, lambdaL2=0)

# close form
W_cf = inv(trainX.T.dot(trainX)).dot(trainX.T.dot(trainY))
```

```python
# Test
print(testX.shape)
testX = np.concatenate((np.ones((testX.shape[0], 1)), testX), axis=1) #(240, 163) axis=1表示左右拼接，axis=0表示上下拼接
print(testX.shape)

# gradient descent
predictedY_gd = np.dot(testX, W_gd)

# Adagrad
predictedY_ada = np.dot(testX, W_ada)

# close form
predictedY_cf = np.dot(testX, W_cf)
```

```python
###---Visualization---###
plt.plot(np.arange(len(list_cost_gd[3:])), list_cost_gd[3:], "b--", label="GD_0")
plt.plot(np.arange(len(list_cost_gd_1[3:])), list_cost_gd_1[3:], "r--", label="GD_100")
plt.plot(np.arange(len(list_cost_ada[3:])), list_cost_ada[3:], "g--", label="Adagrad")
plt.title("Train Process")
plt.xlabel("Iteration")
plt.ylabel("Cost Function (MSE)")
plt.legend()
plt.savefig(os.path.join(os.path.dirname('__file__'), "output_01/TrainProcess"))
plt.show()

# compare predict value with different methods
dcitD = {"Adagrad":predictedY_ada, "CloseForm":predictedY_cf, "GD":predictedY_gd}
pdResult = pd.DataFrame(dcitD)
pdResult.to_csv(os.path.join(os.path.dirname('__file__'), "output_01/Predict"))
print(pdResult)

# visualize predict value with different methods
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.plot(np.arange(len(predictedY_ada)), predictedY_ada, "b--")
plt.title("Adagrad")
plt.xlabel("Test Data Index")
plt.ylabel("Predict Result")
plt.subplot(132)
plt.plot(np.arange(len(predictedY_cf)), predictedY_cf, "r--")
plt.title("CloseForm")
plt.xlabel("Test Data Index")
plt.ylabel("Predict Result")
plt.subplot(133)
plt.plot(np.arange(len(predictedY_gd)), predictedY_gd, "g--")
plt.title("GD")
plt.xlabel("Test Data Index")
plt.ylabel("Predict Result")
plt.tight_layout()
plt.savefig(os.path.join(os.path.dirname('__file__'), "output_01/Compare"))
plt.show()
```
